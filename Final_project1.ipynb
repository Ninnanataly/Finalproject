```python
import spacy
import dacy
import pandas as pd
```

```python
nlp = dacy.load("da_dacy_large_trf-0.1.0")
```

```python
DRB_FULL = pd.read_csv("NinnaNatalyFriis#1429/DRB_FULL.csv")
```

```python
DRB_FULL["Tekst"]= DRB_FULL["Tekst"].astype(str)
```

```python
docs = list(nlp.pipe(DRB_FULL.Tekst))
```

```python
per = []
docid = []

for idi, i in enumerate(docs):
    for entity in i.ents:
        if entity.label_ == 'PER':
            per.append(entity)
            docid.append(idi)
```

```python
DRB_names = pd.DataFrame ()
```

```python
DRB_names['doc_id'] = docid
DRB_names['names'] = per
```
